```python
# Importar las librerías necesarias
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForCausalLM

# Cargar el dataset ConvAI2 Persona-Chat
dataset = load_dataset("conv_ai_2", trust_remote_code=True)

# Mostrar un resumen del dataset
print("Resumen del Dataset:")
print(dataset)

# Mostrar los primeros ejemplos del dataset de entrenamiento
print("\nPrimer Ejemplo del Dataset de Entrenamiento:")
train_example = dataset['train'][0]

# Mostrar las claves disponibles en el ejemplo para entender la estructura
print("Claves disponibles en el ejemplo:", train_example.keys())

# Acceder a los datos relevantes del ejemplo
print("\nID del Diálogo:", train_example['dialog_id'])
print("Diálogo:", train_example['dialog'])
print("Perfil del Bot:", train_example['bot_profile'])
print("Perfil del Usuario:", train_example['user_profile'])
print("Puntuación de Evaluación:", train_example['eval_score'])
print("Compatibilidad de Perfil:", train_example['profile_match'])
```

```python
def preprocess_function(examples):
    inputs = []
    for dialog, bot_profile, user_profile in zip(examples['dialog'], examples['bot_profile'], examples['user_profile']):
        bot_profile_text = " ".join(["".join(sublist) for sublist in bot_profile])
        user_profile_text = " ".join(["".join(sublist) for sublist in user_profile])
        dialog_text = " ".join([turn['text'] for turn in dialog])
        input_text = " ".join([bot_profile_text, user_profile_text, dialog_text])
        inputs.append(input_text)
    return {"inputs": inputs}

# Aplicar el preprocesamiento a todo el dataset
dataset = dataset.map(preprocess_function, batched=True)
print("Datos preprocesados:")
print(dataset['train'][0]['inputs'])

```

```python
from transformers import AutoTokenizer

# Cargar el tokenizador
tokenizer = AutoTokenizer.from_pretrained("microsoft/DialoGPT-small")

# Verificar y establecer el token de padding si no está definido
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# Función de tokenización
def tokenize_function(examples):
    return tokenizer(examples['inputs'], padding=True, truncation=True, max_length=512)

# Aplicar la tokenización al dataset
tokenized_dataset = dataset.map(tokenize_function, batched=True)

print("Datos tokenizados:")
print(tokenized_dataset['train'][0])

```

```python
from transformers import Trainer, TrainingArguments, AutoModelForCausalLM

# Cargar el modelo de lenguaje preentrenado
model = AutoModelForCausalLM.from_pretrained("microsoft/DialoGPT-small")

# Definir los argumentos de entrenamiento
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="steps",
    eval_steps=500,
    logging_dir="./logs",
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    num_train_epochs=3,
    save_steps=10,
    warmup_steps=500,
    weight_decay=0.01,
    logging_steps=100,
    learning_rate=5e-5,
    fp16=True,
    push_to_hub=False
)

# Añadir las etiquetas (labels) para calcular la pérdida
def add_labels(examples):
    examples['labels'] = examples['input_ids'].copy()
    return examples

tokenized_dataset = tokenized_dataset.map(add_labels, batched=True)

# Inicializar el trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset['train'],  # Usar la división 'train' del dataset
    eval_dataset=tokenized_dataset['train'],  # Usar la división 'train' para evaluación temporal (debería ser un conjunto de validación separado)
)

# Ajuste fino del modelo
trainer.train()

# Guardar el modelo ajustado
trainer.save_model("./adjusted_model")

```
